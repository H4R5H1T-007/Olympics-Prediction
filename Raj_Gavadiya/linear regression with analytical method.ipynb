```python
import pandas as pd
import numpy as np

df = pd.read_csv('wheat_production_in_past_years.csv')
```

```python
X_Mean = df['year(X)'].mean()
Y_Mean = df['Production(Y)'].mean()

#print(" yearMean" , yearMean , "\n" , "productionMean" , productionMean)
```

```python
df['U = X - !X'] =df['year(X)']- X_Mean 
df['V = Y - !Y'] =df['Production(Y)'] - Y_Mean
#df
```

```python
df['U*U'] = df['U = X - !X']**2
df['P*P'] = df['Production(Y)']**2
df['U*V'] = df['U = X - !X']*df['V = Y - !Y']
```

```python
UV_Mean = df['U*V'].mean()
UV_Mean
```

```python
variance = df['U*U'].sum() / 48
covariance = df['U*V'].sum() / 48 
print("covariance = " , covariance)
print("variance = " , variance)
```

```python
#if we need line of regression for our prediction so we need wquation  Y = W1*X + W0

W1 = covariance / variance
W0 = Y_Mean - (W1*X_Mean)

# so our line of regression is :::

df['prediction'] = W1* df['year(X)'] + W0
```
