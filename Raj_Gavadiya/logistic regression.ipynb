```python
import numpy as np
import pandas as pd

df = pd.read_excel('Book2.xlsx')

arr = df.to_numpy()
X = (arr[: , 0])
Y = (arr[: , 1])
m = len(X)
X = np.reshape(X , (len(X),1))
Y = np.reshape(Y ,(len(X) ,1) )
X = (X - 18)/(62-18)
X = np.c_[np.ones((len(X),1)),X]
#Y = (Y -9854)/(80679 -9854)
m=len(X)

rate = 0.723

iter = 1000
theta_best = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
print("theta by analytical method : " , theta_best.T)
#theta = np.random.randn(2,1)
theta = np.array ([[0], [0]])
for i in range(iter):        
         prediction = np.dot(X,theta)
        # print(prediction)
         theta = theta -(rate/len(X))*( np.dot(X.T , (prediction - Y)))
        # print(theta.T)
print(theta , rate)
```
