```python
# import libraries
import numpy as np
import pandas as pd

# import data 
df = pd.read_excel('wheat_production.xlsx')
```

```python

#convert data in array
arr = df.to_numpy()

# assign columns to variable X and Y
X = (arr[: , 0])
Y = (arr[: , 1])

# reshhape column matrix that we need
X = np.reshape(X ,(len(X), 1) )
Y = np.reshape(Y ,(len(X) ,1) )

#normalize datA in range 0 to 1`......
X = (X - 1960)/(2009 -1960)
Y = (Y -9854)/(80679 -9854)

#  made a matrix of X0 and X1 by combining of X and other column of 1
X = np.c_[np.ones((len(X),1)),X]

#print few rows of X matrix::::::
 
X[0: 10]
```

```python
# we also predict  by analytical method to compare both prediction method 

theta_best = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
print(" Theta[0] :", theta_best[0])
print(" Theta[1] :", theta_best[1])
```

```python
m=len(X)

#decide rate and iterations
rate =0.723
iter = 100

#dedcide initial value of theta matrix to random....
theta = np.random.randn(2,1)

for i in range(iter):        
        prediction = np.dot(X,theta)
       # print(prediction)
        theta = theta -(rate/m)*( np.dot(X.T , (prediction - Y)))
       # print(theta.T)

print(" Theta[0] :", theta[0])
print(" Theta[1] :", theta[1])
```

```python
# assign all predicted value to new variable by dot product of two matrix( theta , X)
rr = np.dot(X,theta)

#converting into its original form
Y = rr * (80679 -9854) + 9854

#add a new column in our data to compare both production 
df['predicted_production'] = Y
```
